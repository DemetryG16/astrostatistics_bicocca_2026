# Probability Desity Function (PDF)
---

### The pdf is a function that tells you the probability of having a value between a set of values:
* **Example:** If you have the pdf f(x):

     Take $x=9$. **The the probablity of having the value $x=9$ is f(9) out of this set of data.**

#### Follow this example to understand:

**!! This example shows the Probability Mass Function (PMF). But it is the same principle for a pdf.**

```python
# You have this data set:
data_set = [1,2,3,0.5,0.55,0.6,1,1,1,5,0,7,7,7,7,4,20,20,20,20,20,20,20,20,19,18,17]


freq = np.ones_like(data_set)*100/len(data_set) # The percentage of getting a value.

fig , (ax1,ax2,ax3,ax4) = plt.subplots(1,4,figsize= (12,5))
ax1.hist(data_set, bins= 100, color = 'green')
ax1.set_ylabel("The number of each value")
ax1.set_title('Simple Histogram')
ax1.set_xlabel("x")

ax2.hist(data_set, bins= 100, weights= freq, color = 'red')
ax2.set_ylabel("Probability %")
ax2.set_title('The percentage of each')
ax2.set_xlabel("x")

# For ax3, get each value with its counts probability to find the probability function.
values, counts = np.unique(data_set, return_counts=True)
prob = counts / len(data_set) * 100

ax3.plot(values, prob, color = 'blue')
ax3.set_ylabel("Probability %")
ax3.set_xlabel("x")
ax3.set_title('PMF = f(x) ~ PDF')

ax4.plot(values, prob, color = 'blue')
ax4.hist(data_set, bins= 100, weights= freq, color = 'red')
ax4.set_ylabel("Probability %")
ax4.set_xlabel("x")
ax4.set_title('Combined')


fig.suptitle('Transfrom histogram into a function')
fig.tight_layout()
plt.show()
```
